```python
import numpy.typing as npt 

# data for x axis
x: npt.ArrayLike = [1,2,3,4,5,6,7]

# data for y axis
y1: npt.ArrayLike = [7,6,5,4,3,2,1]
```

```python
import matplotlib.pyplot as plt

plt.plot(x, y1)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

```python
y2: npt.ArrayLike = [w*w for w in x]
plt.plot(x,  y1)
plt.plot(x, y2)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

```python
import math
y3: npt.ArrayLike = [math.log(w) for w in x]
plt.plot(x,  y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

```python
y4: npt.ArrayLike = [math.exp(w) for w in x]
plt.plot(x,  y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.plot(x, y4)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```
